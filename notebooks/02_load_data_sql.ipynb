```python
import pandas as pd

file_add = "../data/raw/Online_Retail.xlsx"
df_retail = pd.read_excel(file_add, sheet_name="Online Retail") 

df_retail
```

```python

df_invoices = (
    df_retail.groupby("InvoiceNo", as_index=False)
    .agg({
        "InvoiceDate" : "min",
        "CustomerID" : "first",
        "Country" : "first"
    })
)

df_invoices = df_invoices.rename(columns={
    "InvoiceNo" : "invoice_no",
    "InvoiceDate" : "invoice_date",
    "CustomerID" : "customer_id",
    "Country" : "country"
})


df_invoiceitems = (
    df_retail[
        ["InvoiceNo", "StockCode", "Description", "Quantity", "UnitPrice"]
    ]
).copy()

df_invoiceitems = df_invoiceitems.rename(columns={
    "InvoiceNo" : "invoice_no",
    "StockCode" : "stock_code",
    "Description" : "description",
    "Quantity" : "quantity",
    "UnitPrice" : "unit_price"
})
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "postgresql+psycopg://postgres:Patriotism0102@localhost:5432/retail_sales"
)

df_invoiceitems.to_sql(
    "invoiceitems",
    engine,
    if_exists = "append",
    index=False
)
```
