# **Exploratory Data Analysis**

## **Dataset Information**
This online retail transaction dataset was extracted from the UC Irvine Machine Learning Repository. It contains all the **541909 transactions** occuring for a UK-based and registered, non-store online retail between 01/12/2009 and 09/12/2011. The company mainly seels unique all-occasion gift-ware. Many customers of the company are wholesalers



### **Dataset Structure**

Each row represents an invoice item, with their corresponding:

    - InvoiceNo
    - StockCode
    - Description
    - Quantity
    - InvoiceData
    - UnitPrice
    - CustomerID
    - Country

```python
import pandas as pd

file_add = "../data/raw/Online_Retail.xlsx"
df_retail = pd.read_excel(file_add, sheet_name="Online Retail") 

df_retail
```

### **StockCode**

Represents the unique identifier of the product.

There are special stock codes that do not represent actual physical prodcuts. They represent operational transcations:

Some Examples:

    - DOT
    - POST
    - BANK CHARGES
    - M
    - D
    - C2
    - CRUK


Another special `StockCode` are codes that start with '**C**', and it seems that all entries that satisfies this, has a negative quantity. This implies that entries with `StockCode` that starts with C are **returned items**.

```python
len_start_C = len(df_retail[df_retail['InvoiceNo'].astype(str).str.startswith("C")]["Quantity"])
len_start_C_neg_quant = len(
     df_retail
    [
    (df_retail['InvoiceNo'].astype(str).str.startswith("C")) & 
    (df_retail["Quantity"] <=0)
    ]
)

print(f"{(len_start_C_neg_quant/len_start_C)*100}% of entries with StockCode that starts with C has a negative quantity.")
```

### **Description**

This contains a short description of the product.

Ideally, there should be a one-to-one correspondence between `StockCode` and `Description`. However, there seems to be some exceptions, which we will investigate.




```python
print("Number of StockCodes with X duplicates, where X is the first column(1,2,0,3,...,7)")
df_retail.groupby("StockCode")['Description'].nunique().value_counts()
```

It appears that the `Description` column is also being used to record additional information about a product, rather than serving as a unique product name. As a result, some `StockCode`'s are associated with multiple unique descriptions.

The code below identifies all `StockCode`'s that are associated with **N unique descriptions** , allowing us to inspect these cases individually and determine whether the differences are due to data entry inconsistencies, product variations, or other reasons.

```python
#N stands for how many duplicate description
N = 5

stockcodes_2 = (
    df_retail.groupby("StockCode")["Description"]
    .nunique()
    .loc[lambda x: x == N]
)


stockcodes_list = df_retail[df_retail["StockCode"].isin(stockcodes_2.index)]["StockCode"].to_list()

print("Duplicate Descriptions of StockCode:", stockcodes_list[0])
df_retail[df_retail['StockCode'] == stockcodes_list[0]]['Description'].value_counts()
```

In this column, there are 1454(*0.26%*) missing values and it seems that all of these entries have:

    - CustomerID = NaN
    - UnitPrice = 0

Since these rows have a unit price of zero, they do not contribute to revenue calculations. Furthermore, the missing product descriptions make it difficult to determine what these records represent. 

At this stage, these rows appear to be incomplete or administrative records rather than valid sale transactions. They will be investigated further before deciding whether to remove them during data cleaning.

```python
len_nan_customer_desc=len(df_retail[df_retail['Description'].isna() & df_retail['CustomerID'].isna() & (df_retail["UnitPrice"] == 0.0)])
len_nan_desc = len(df_retail[df_retail['Description'].isna()])

print(f"{(len_nan_customer_desc/len_nan_desc)*100}% of entries with NaN StockCode has NaN CustomerID AND 0 UnitPrice")

print("No. of invoices:", )
df_retail[df_retail['Description'].isna() & df_retail['CustomerID'].isna() & (df_retail["UnitPrice"] == 0.0)]

```

Furthermore, all entries with NaN `CustomerID`, `Description` and 0 `UnitPrice` only has one invoice item. The code below prints the number of unique `InvoiceNo` in the table above. The number of rows matched means there were no duplicate `InvoiceID`(invoices with multiple items).

```python
df_retail[df_retail['Description'].isna() & df_retail['CustomerID'].isna() & (df_retail["UnitPrice"] == 0.0)]['InvoiceNo'].nunique()
```

```python

df_invoices = (
    df_retail.groupby("InvoiceNo", as_index=False)
    .agg({
        "InvoiceDate" : "min",
        "CustomerID" : "first",
        "Country" : "first"
    })
)

df_invoices = df_invoices.rename(columns={
    "InvoiceNo" : "invoice_no",
    "InvoiceDate" : "invoice_date",
    "CustomerID" : "customer_id",
    "Country" : "country"
})


df_invoiceitems = (
    df_retail[
        ["InvoiceNo", "StockCode", "Description", "Quantity", "UnitPrice"]
    ]
).copy()

df_invoiceitems = df_invoiceitems.rename(columns={
    "InvoiceNo" : "invoice_no",
    "StockCode" : "stock_code",
    "Description" : "description",
    "Quantity" : "quantity",
    "UnitPrice" : "unit_price"
})
```
