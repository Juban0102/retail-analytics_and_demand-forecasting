# Exploratory Data Analysis

## Dataset Information
This online retail transaction dataset was extracted from the UC Irvine Machine Learning Repository. It contains all the **541909 transactions** occuring for a UK-based and registered, non-store online retail between 01/12/2009 and 09/12/2011. The company mainly seels unique all-occasion gift-ware. Many customers of the company are wholesalers



### Dataset Structure

Each row represents an invoice item, with their corresponding:

    - InvoiceNo
    - StockCode
    - Description
    - Quantity
    - InvoiceData
    - UnitPrice
    - CustomerID
    - Country

```python
import pandas as pd

file_add = "../data/raw/Online_Retail.xlsx"
df_retail = pd.read_excel(file_add, sheet_name="Online Retail") 

df_retail
```

### Data Anomalies



**InvoiceNo**

Represents the unique identifier for each invoice. 
For each invoices, you can have multiple items. That's why you can see duplicate InvoiceNo.

No missing values.

```python
df_retail.info()
```

**StockCode**

Represents the unique identifier of the product.

There are special stock codes that do not represent actual physical prodcuts. They represent operational transcations:

Some Examples:

    - DOT
    - POST
    - BANK CHARGES
    - M
    - D
    - C2
    - CRUK


**Description**

This contains a short description of the product.

**CustomerID**

Approximately 25% of total transactions **(or 135080 transactions)** have a missing `CustomerID`

```python
df_retail.info()
```

**UnitPrice**

There are no missing values for `UnitPrice`, but there are **2515 entries** with 0.00 UnitPrice values which is an anomaly(what products costs 0 each unit?). 
Maybe these are promotional gifts, buy-one-get-one free items, samples, replacements, internal inventory transfers, etc.. 

*No action taken until we gather more information.*





```python
df_retail[df_retail["UnitPrice"]==0]
```


Also, 2475 entries out of the total 2515 entries with 0 `UnitPrice` have **missing `CustomerID`**.

```python
len_zero_unitprice = len(df_retail[df_retail['UnitPrice'] == 0 ])
len_customer_id_null = len(df_retail[(df_retail['CustomerID'].isnull()) & (df_retail['UnitPrice'] == 0 )])

print(f"Number of rows with zero UnitPrice: {len_zero_unitprice}")
print(f"{len_customer_id_null/len_zero_unitprice*100:.2f}% of rows with zero UnitPrice have null CustomerID")
```

```python

df_invoices = (
    df_retail.groupby("InvoiceNo", as_index=False)
    .agg({
        "InvoiceDate" : "min",
        "CustomerID" : "first",
        "Country" : "first"
    })
)

df_invoices = df_invoices.rename(columns={
    "InvoiceNo" : "invoice_no",
    "InvoiceDate" : "invoice_date",
    "CustomerID" : "customer_id",
    "Country" : "country"
})


df_invoiceitems = (
    df_retail[
        ["InvoiceNo", "StockCode", "Description", "Quantity", "UnitPrice"]
    ]
).copy()

df_invoiceitems = df_invoiceitems.rename(columns={
    "InvoiceNo" : "invoice_no",
    "StockCode" : "stock_code",
    "Description" : "description",
    "Quantity" : "quantity",
    "UnitPrice" : "unit_price"
})
```
